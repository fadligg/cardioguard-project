```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv('../../data/processed/cardio_cleaned.csv')
df.head()
```

```python
print("Jumlah Baris dan Kolom:", df.shape)
print("\nTipe Data:")
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
sns.countplot(x='cardio', data=df)
plt.title("Distribusi Risiko Penyakit Kardiovaskular")
plt.xlabel("Cardio (0=Tidak, 1=Ya)")
plt.ylabel("Jumlah")
plt.show()
```

```python
sns.histplot(data=df, x='age_years', hue='cardio', bins=30, kde=True, multiple="stack")
plt.title("Distribusi Umur Berdasarkan Risiko Cardio")
plt.xlabel("Umur")
plt.show()
```

```python
sns.boxplot(x='cardio', y='bmi', data=df)
plt.title("Perbandingan BMI terhadap Risiko Cardio")
plt.show()
```

```python
sns.boxplot(x='cardio', y='ap_hi', data=df)
plt.title("Tekanan Darah Sistolik vs Cardio")
plt.show()
```

```python
sns.countplot(x='cholesterol', hue='cardio', data=df)
plt.title("Kolesterol terhadap Risiko Cardio")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Korelasi Antar Variabel")
plt.show()
```

```python
corr = df.corr()['cardio'].sort_values(ascending=False)
print(corr)
```
